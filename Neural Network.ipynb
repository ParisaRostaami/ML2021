1.load Data

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
###########################
import tensorflow as tf
from tensorflow import keras
from keras.models import Sequential
from keras.layers import Dense
```

```python
link = 'https://raw.githubusercontent.com/jbrownlee/Datasets/master/pima-indians-diabetes.data.csv'
dataset = pd.read_csv(link, header=None)
```

```python
dataset.head()
```

```python
dataset.columns = ['num_preg','glucose','blood pressure','Triceps skin','serum insulin','Body mass','Diabetes pedigree','Age','y']
```

```python
dataset.head()
```

```python
x=dataset.drop(['y'],axis=1)
y=dataset['y']
```

```python
x
```

```python
!pip install -U scikit-learn
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=40)
```

2.Define Kears Model

```python
model = Sequential()
model.add(Dense(6, input_dim=8, activation='relu'))
model.add(Dense(4, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
```

```python
# model = keras.Sequential([
#     keras.layers.Dense(6, input_dim=8, activation='relu'),
#     keras.layers.Dense(4, activation='relu'),
#     keras.layers.Dense(2, activation='relu'),
#     keras.layers.Dense(1, activation='sigmoid')
# ])
```

3.compile keras model

```python
model.compile(loss = 'binary_crossentropy',
              optimizer = 'adam',
              metrics = ['accuracy'])
```

```python
print(model.summary())
```

4.fit keras model

```python
model.fit(x_train, y_train, 
          epochs = 80, 
          batch_size = 2,
          verbose = 1)
```

5. Evaluate Keras Model

```python
score = model.evaluate(x_test, y_test, verbose=1)
```

```python
score
```

```python
print('Test loss:', score[0])
print('Test accuracy:', score[1])
```

```python
history = model.fit(x, y, validation_split=0.3, epochs=50, batch_size=10, verbose=0)
```

```python
print(history.history.keys())
```

```python
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper left')
plt.show()
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper left')
plt.show()
```

```python
model.save("model.h5")
print("Saved model to disk")
```

```python
model = tf.keras.models.load_model('model.h5')
model.summary()
```

```python
score = model.evaluate(x_test, y_test, verbose=1)
print("%s: %.2f%%" % (model.metrics_names[1], score[1]*100))
```
